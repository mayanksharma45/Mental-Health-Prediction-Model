### Treating/Handling Missing Values

```python
#to get rid of the variables "Timestamp",“comments”, “state” as these attributes are either string values or date values

data = data.drop(['comments'], axis= 1)    # it having a lot of missing values
data = data.drop(['state'], axis= 1)
data = data.drop(['Timestamp'], axis= 1)
```

```python
data.head(5)
```

```python
# Assign default values for each data type
defaultInt = 0
defaultString = 'NaN'

# Create lists by data tpe
Num_Features = ['Age']
Cat_Features = ['Gender', 'Country', 'self_employed', 'family_history', 'treatment', 'work_interfere',
                 'no_employees', 'remote_work', 'tech_company', 'anonymity', 'leave', 'mental_health_consequence',
                 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
                 'mental_vs_physical', 'obs_consequence', 'benefits', 'care_options', 'wellness_program',
                 'seek_help']

for feature in data:
    if feature in Num_Features:
        data[feature] = data[feature].fillna(defaultInt)
    elif feature in Cat_Features:
        data[feature] = data[feature].fillna(defaultString)
    else:
        print('Error: Feature %s not found.' % feature)
```

### Standardizing Gender labels 

```python
gender = data['Gender'].str.lower() #lower case all column's elements
```

```python
gender = data['Gender'].unique() #Select unique elements

# Making gender groups
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]
trans_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", 
             "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]           
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

for (row, col) in data.iterrows():

    if str.lower(col.Gender) in male_str:
        data['Gender'].replace(to_replace=col.Gender, value='male', inplace=True)

    if str.lower(col.Gender) in female_str:
        data['Gender'].replace(to_replace=col.Gender, value='female', inplace=True)

    if str.lower(col.Gender) in trans_str:
        data['Gender'].replace(to_replace=col.Gender, value='trans', inplace=True)

# Removing invalid values
stk_list = ['A little about you', 'p']
data = data[~data['Gender'].isin(stk_list)]    # means it select which row of Gender that is not in stk_list

print(data['Gender'].unique())
```

### Handling Outliers

```python
# checking quantiles
q1 = data['Age'].quantile(0.25)   # first quartile, q1
q3 = data['Age'].quantile(0.75)   # third quartile, q3

#calculating iqr
iqr = q3 - q1   # inter-quartile range, iqr

# Calculation upper_limit and lower_limit
upper_limit = q3 + 1.5*iqr   
lower_limit = q1 - 1.5*iqr
upper_limit, lower_limit

# imputing outliers by defing a function
def limit_imputer(value):
  if value > upper_limit:
    return upper_limit
  if value < lower_limit:
    return lower_limit
  else:
    return value
```

```python
data['Age'] = data['Age'].apply(limit_imputer)
```

```python
#Replace "NaN" string with 'No'
data['self_employed'] = data['self_employed'].replace([defaultString], 'No')
print(data['self_employed'].unique())
```

### Visulisation of Variables

```python
def bar_plt(df, x, y, tit):
    plt.figure(figsize=(12, 6))
    a= sns.barplot(data= df, x=x, y=y)
    for i in a.containers:
        a.bar_label(i)
    plt.title(tit.upper())
    plt.xlabel(x.upper())
    plt.gca().get_yaxis().set_visible(False)
    plt.show()
```

```python
def pie_plt(x, lab, tit):
    plt.figure(figsize=(8, 4))
    plt.pie(x, labels= lab, autopct = '%1.1f%%', textprops={'fontsize': 8})
    plt.title(tit.upper())
    plt.show()
```

```python
def val_cnts(df, col):
    return df[col].value_counts().reset_index()
```

### Encoding the data

```python
from sklearn.preprocessing import LabelEncoder
labelDictionary = {}
for feature in data:
    le = LabelEncoder()
    le.fit(data[feature])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    data[feature] = le.transform(data[feature])
    # Get labels
    labelKey = 'label_' + feature
    labelValue = [*le_name_mapping]
    labelDictionary[labelKey] =labelValue
    
for key, value in labelDictionary.items():     
    print(key, value)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

X = data.drop(columns=['treatment'])  # Excluir la variable objetivo
y = data['treatment']
extra_trees = ExtraTreesClassifier(random_state=42)
extra_trees.fit(X, y)

feature_importances = extra_trees.feature_importances_

extra_trees_results = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Mostrar las características más importantes
print("Importance of characteristics using ExtraTreesClassifier:")
print(extra_trees_results)
```

```python
# Visualización de las importancias de características
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.barh(extra_trees_results['Feature'], extra_trees_results['Importance'], color='blue')
plt.xlabel('Importancia')
plt.title('Importance of characteristics using ExtraTreesClassifier')
plt.gca().invert_yaxis()
plt.show()
```

- Work Interference is the __Dominant Predictor__ for treatment
- Family History is the Second Most Important
